### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
municipios_granada = dataset[dataset['Provincia'] == 'Granada']
poblacion_granada = municipios_granada['Poblacion']
media_granada = poblacion_granada.mean()
mediana_granada = poblacion_granada.median()
desviacion_estandar_granada = poblacion_granada.std()
maximo_granada = poblacion_granada.max()
minimo_granada = poblacion_granada.min()
print("Media:", media_granada)
print("Mediana:", mediana_granada)
print("Desviación Estándar:", desviacion_estandar_granada)
print("Máximo:", maximo_granada)
print("Mínimo:", minimo_granada)
```

### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
plt.figure(figsize=(12, 8))
for provincia in dataset['Provincia'].unique():
    poblacion_provincia = dataset[dataset['Provincia'] == provincia]['Poblacion']
    plt.hist(poblacion_provincia, alpha=0.5, label=provincia)
plt.xlabel('Población')
plt.ylabel('Frecuencia')
plt.title('Histograma de Población por Provincia')
plt.show()
```

### 7) Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
import random

m_al = random.sample(list(dataset['Municipio'].unique()), 50)
m_al_dt = dataset[dataset['Municipio'].isin(m_al)]
prom_al = m_al_dt['Poblacion'].mean()
desvt_al = m_al_dt['Poblacion'].std()
print("Promedio de población de los 50 municipios aleatorios:", prom_al)
print("Desviación estándar de población de los 50 municipios aleatorios:", desvt_al)
```

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
madrid = dataset[dataset['Municipio'] == 'Madrid'].iloc[0]
x_madrid = madrid['X']
y_madrid = madrid['X']
dataset['Distancia_a_Madrid'] = np.sqrt((dataset['X'] - x_madrid)**2 + (dataset['Y'] - y_madrid)**2)
datos_sin_madrid = dataset[dataset['Municipio'] != 'Madrid']
municipio_cercano_madrid = datos_sin_madrid.loc[datos_sin_madrid['Distancia_a_Madrid'].idxmin()]
municipio_lejano_madrid = datos_sin_madrid.loc[datos_sin_madrid['Distancia_a_Madrid'].idxmax()]
print("Municipio más cercano a Madrid:", municipio_cercano_madrid['Municipio'])
print("Municipio más lejano a Madrid:", municipio_lejano_madrid['Municipio'])
```
